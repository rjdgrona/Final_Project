```python
# importing required libraries
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
```

```python
df = pd.read_csv('data/FinalDataset.csv', parse_dates=["DateTime"])
station = df[df['Station']=='COLS'].reset_index(drop=True)
station.head()
```

```python
station_plot = station[['DateTime','Origin']].set_index('DateTime')
plt.figure(figsize=(15,6))
plt.plot(station_plot)
plt.title("Per Station passenger trend (Jan 2016 ~ May 2017)")
plt.savefig('images/daily_COLS.png')
plt.show()
```

```python
#creating dataframe
new_data = pd.DataFrame(index=range(0,len(df)),columns=['Date', 'Origin'])

for i in range(0,len(station)):
    new_data['Date'][i] = station['DateTime'][i]
    new_data['Origin'][i] = station['Origin'][i]
    
new_data.head()
```

```python
#setting index
new_data.index = new_data.Date
new_data.drop('Date', axis=1, inplace=True)
new_data.head()
```

```python
new_data = new_data["2017"]
print(new_data.head())
new_data
new_data.size
```

```python
#creating train and test sets
dataset = new_data.values

print(dataset.size)

train = dataset[0:2000,:]
valid = dataset[2000:,:]

print(train.size)
print(valid.size)
```

```python
#converting dataset into x_train and y_train
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)

x_train, y_train = [], []
for i in range(60,len(train)):
    x_train.append(scaled_data[i-60:i,0])
    y_train.append(scaled_data[i,0])
x_train, y_train = np.array(x_train), np.array(y_train)

x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

```python
# create and fit the LSTM network
model = Sequential()

model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.2))

model.add(LSTM(units=50))
model.add(Dropout(0.2))

model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mean_squared_error')
model.fit(x_train, y_train, epochs=100, batch_size=24)
```

```python
#predicting future values, using past 60 from the train data
inputs = new_data[len(new_data) - len(valid) - 60:].values
inputs = inputs.reshape(-1,1)
inputs  = scaler.transform(inputs)

X_test = []
for i in range(60,inputs.shape[0]):
    X_test.append(inputs[i-60:i,0])
X_test = np.array(X_test)

X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
riders = model.predict(X_test)
riders = scaler.inverse_transform(riders)
```

```python
rms=np.sqrt(np.mean(np.power((valid-riders),2)))
rms
```

```python
#for plotting the results overall

train = new_data[:2000]
valid = new_data[2000:]
plt.figure(figsize=(15,6))
valid['Predictions'] = riders
plt.plot(train['Origin'], color='blue', label='Train COLS', alpha=.5)
plt.plot(valid['Origin'], color='orange', label='Test COLS')
plt.plot(valid['Predictions'], color='green', label='Predictions COLS',alpha=.5)
plt.title('Train and Test COLS BART data in 2017')
plt.xlabel('Date')
plt.ylabel('# of riders')
plt.legend()
plt.savefig("images/train_test_prediction_cols.png", dpi=150)
```
