```python
sns.set_context("notebook", font_scale=1.)
sns.set_style("whitegrid")
%config InlineBackend.figure_format = 'retina'
```

```python
Stations = pd.read_csv("data/FinalDataset.csv", header=0, parse_dates=["DateTime"], squeeze=True)
```

```python
Stations = Stations[Stations['Station']=='COLS'].reset_index(drop=True)

df = Stations[Stations["Year"]==2017]

df=df[["DateTime","Origin"]].set_index("DateTime")
df.tail()
```

```python
df.info()
```

```python
df.plot(color='red',figsize=(25,8))
plt.title("COLS - Daily passenger trend (Jan ~ May 2017)")
plt.legend(loc="best")
plt.savefig('images/COLS_actual_daily.png')
```

```python
# Fit the first model.
model = ARIMA(df, order=(2,0,1))
model_fit = model.fit(disp=0)
model_fit.summary()
```

```python
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
residuals.plot(kind='kde')
plt.savefig('images/test_arima_COLS.png')
print(residuals.describe())
```

```python
X = df.values
Y = df.index.values
size = int(len(X) * 0.70)
train, test = X[0:size], X[size:len(X)]
y_train, y_test = Y[0:size], Y[size:len(X)]
history = [x for x in train]
predictions = list()

for t in range(len(test)):
    model = ARIMA(history, order=(2,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
    
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
```

```python
# LOOKING AT Total planned 
fig = plt.figure(figsize=(25, 8))
# plot
plt.plot(y_test,test)
plt.plot(y_test, predictions, color='red')
plt.savefig('images/arima_COLS.png')
plt.show()
```
